```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins=sns.load_dataset('penguins')
penguins.head(3)
```

```python
sns.set_theme()
sns.histplot(data=penguins,x="body_mass_g",bins=20)
```

```python
sns.kdeplot(data=penguins,x="body_mass_g")
```

```python
sns.kdeplot(data=penguins,x="body_mass_g", hue='species',multiple='stack')
```

```python
sns.histplot(data=penguins, x='flipper_length_mm')
```

```python
sns.histplot(data=penguins,x='body_mass_g', y='flipper_length_mm')
```

몸무게는 3700-4000 날개 길이는 190-195사이에 펭귄이 가장 많은 펭귄이 분포되어 있음

```python
sns.kdeplot(data=penguins,x='body_mass_g', y='flipper_length_mm')
```

```python
sns.kdeplot(data=penguins,x='bill_length_mm', y='flipper_length_mm',hue='species',multiple='stack')
```

x는 부리의 길이, y는 날개의 길이를 의미한다.
Adelie가 전반적으로 부리와 날개의 길이가 작고,
Gentoo가 전반적으로 날개의 길이가 크고,
Chinstrap가 전반적으로 부리 길이가 크다,
Adelie는 부리 35-40사이, 날개 185-193,
Chinstrap은 부리 48-53, 날개 195-200,
Gentoo는 부리 45-51, 날개 202-215 정도에서 데이터가 밀집되어 있다.

세가지 펭귄 종 중에서 Gentoo가 부리길이와 날개길이 관계가 가장 정비례에 가깝다. 즉, 부리가 길 때, 날개가 길다고 표현하기 가장 적절한 종이다. 

Gentoo의 경우, 부리와 날개의 길이가 짧을 수록 상관관계가 높은 것이 밀집되어 있고, 
Chinstrap의 경우 부리와 날개의 길이가 길수록 상관관계가 높은 것이 밀집되어 있다. 
Adelie의 경우, 전반적으로 펭귄의 부리길이, 날개길이가 서로 대강 연관되어 있다. 

```python
sns.histplot(data=penguins,x='bill_length_mm', y='flipper_length_mm', hue='species')
```

```python
tips=sns.load_dataset('tips')
tips.head(20)
```

```python
sns.rugplot(data=tips, x='tip', height=0.2)
```

rug 카펫 같은 곳에 데이터가 많은면 직사각형이 된다

```python
sns.histplot(data=tips, x='tip')
```

```python
sns.rugplot(data=tips, y='tip', height=0.2)
```

```python
sns.rugplot(data=tips,x='total_bill', height=0.2)
```

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill', height=0.1)
```

```python
x = tips.total_bill.values
```

```python
sns.distplot(x,rug=True)
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
sns.rugplot(data=tips,x='total_bill',y='tip',height=0.03)
```

합계가 10 - 20사이 팁은 1.7 - 4달러 사이에 팁을 주는 사람이 많다

```python
sns.displot(data=penguins,x='body_mass_g',hue='sex',col='species')
```

종별을 기준으로 성별을 나누고 몸무게 값을 나타낸다. 곂친부분은 색깔이 진해진다
펭귄은 수컷이 몸무게가 많이 나간다.
Gentoo라는 종이 몸무게가 가장 많이 나간다.
Adelie라는 종은 개체수가 가장 많다.
Chinstrap라는 종이 개체수가 가장 적고 몸무게도 가장 적게 나간다.
hist 그래프의 단면적이 가장 작다 = 개체수가 가장 적다

```python
sns.displot(data=tips,kind='kde',x='tip',col='time',rug=True)
```

점심보다 저녁에 팁을 주는 사람이 많고 팁도 더 많다

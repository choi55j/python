Plt에 matplotlib 을 저장했고, matplotlib에 있는 plot이라는 메소드를 통해 어떤 상관관계가 있는 그래프를 그릴 것인지를 지정하고, 해당 그래프의 색깔을 지정한다. 
plot(x축에 대입할 값, y축에 대입할 값, ‘라인 컬러’)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,3)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'r--')
plt.subplot(1,3,1)
plt.plot(y,x,'g*-')
# plt.subplot 괄호에 있는 숫자는 행 열 자리
```

plt.subplot(1,2,1)의 의미는, 기존 plot을 1을 기준으로 가로에 두개의 그래프를 넣는다는 의미이고, 그 다음 1은 두개의 그래프 위치 중 첫번째에 plt.plot(x, y, 'r--')  라는 그래프를 넣는다는 의미이다. 

plt.subplot(1,3,1) 은 기존 plot의  범위에 3개의 그래프를 넣는다는 의미이고, 그 그래프 위치 3가지 중 첫번째에 plt.plot(x, y, 'r--') 그래프를 넣는다는 뜻이다. 


```python
plt.subplot(1,4,1)
plt.plot(x,y,'y--')
plt.subplot(1,4,4)
plt.plot(y,x,'b*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.3,0.9])

axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

Figure라는 메소드는 그래프의 크기와 가로 세로 비율을 정할 수 있도록 해준다.
fig.add_axes([왼쪽축에서의 거리, 바닥에서의 거리, 그래프의 폭, 그래프의 높이]) 
위 add_axes 안에 들어갈 수 있는 숫자의 범위는 0부터 1까지 가능 

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.2,0.6,0.5,0.3])
axes.plot(x,y,'b')
axes.set_xlabel("X Label_axes2")
axes.set_ylabel("Y Label_axes2")
axes.set_title("Axes 2 Title")
axes2.plot(y,x,'r')
axes2.set_xlabel("X Label_axes2")
axes2.set_ylabel("Y Label_axes2")
axes2.set_title("Axes 2 Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel("X Label_axes2")
axes1.set_ylabel("Y Label_axes2")
axes1.set_title("Axes 2 Title")

axes2.plot(y,x,'r')
axes2.set_xlabel("X Label_axes2")
axes2.set_ylabel("Y Label_axes2")
axes2.set_title("Axes 2 Title")
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="#FF69B4", alpha=0.5) #half-transparant
ax.plot(x, x+2, color="#FF1493") #RGB hex code
ax.plot(x, x+3, color="#FFB6C1") #RGB hex code
```

```python
fig = plt.figure()
ax = fig.add_axes([0.5,0.5,1.2,0.8])
ax.plot(x, x+1, color="#FF69B4", alpha=0.5, linewidth=0.5) #half-transparant
ax.plot(x, x+2, color="#FFB6C1",linewidth=0.51) #RGB hex code
ax.plot(x, x+3, color="#FF1493",linewidth=0.52) #RGB hex code
ax.plot(x, x+4, color="#EE82EE")
```

Figsize = 그래프의 사이즈를 가로 * 세로 비율로 확정하는 것.

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", lw=1.00)
ax.plot(x, x+4, color="red", lw=2.00)
ax.plot(x, x+5, color="g", lw=2.00)
ax.plot(x, x+6,"g-.", lw=2.00)
ax.plot(x, x+7,"g:", lw=2.00)
# linewidth를 약자로 lw로 써도 사용가능하다 유용한방법
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", lw=1.00)
ax.plot(x, x+4, color="red", lw=2.00)
ax.plot(x, x+5, color="green", lw=3, ls='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", lw=1.00)
ax.plot(x, x+4, color="red", lw=2.00)
ax.plot(x, x+5, color="green", lw=3, ls='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
ax.plot(x, x+8, color="blue", lw=3, ls='-',marker='+')
ax.plot(x, x+9, color="blue", lw=3, ls='-',marker='o')
ax.plot(x, x+10, color="blue", lw=3, ls='-',marker='s')
ax.plot(x, x+11, color="blue", lw=3, ls='--',marker='1')
```

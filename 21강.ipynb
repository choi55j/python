점심에는 남성과 여성의 식사 차이가 거의 없는데
저녁에는 여성보다 남성이 밖에서 더 많은 식사를 한다
남성은 점심보다 저녁에 더 많이 식사를 한다
여성은 점심과 저녁에 식사 차이가 거의 없다
여성은 점심보다 저녁에 흡연자가 더 많다

```python
sns.histplot(data=tips,x='tip',hue='time', multiple="dodge")
```

```python
penguins=sns.load_dataset('penguins')
penguins
```

```python
sns.histplot(data=penguins, x='body_mass_g',hue='species',multiple='stack',bins=20)
```

```python
sns.set_theme()
plt.figure(dpi=200)
sns.histplot(
    data=penguins, 
    x="body_mass_g", 
    bins=20, 
    hue="species", 
    multiple="stack", 
    kde=True
)
plt.str.title("Penguin Body Mass")
```

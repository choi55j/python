```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salaries = pd.read_csv("data/Salaries.csv", low_memory = False)
```

```python
plt.style.use('fivethirtyeight')

salaries["BasePay"]=pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimePay"]=pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"]=pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
df=salaries[["EmployeeName", "BasePay","OvertimePay","OtherPay"]]
df.set_index("EmployeeName", inplace=True)
```

```python
ufos = pd.read_csv("data/nuforc_reports.csv", low_memory = False)
```

```python
ufos['shape'].value_counts()
```

```python
ufos = pd.read_csv("data/nuforc_reports.csv", parse_dates=['date_time', "posted"])
```

```python
ufos[ufos["shape"] == "formation"].date_time.dt.date.value_counts().sort_index().plot(label = "formation")
ufos[ufos["shape"] != "formation"].date_time.dt.date.value_counts().sort_index().plot(label="the others")
plt.legend()
```

```python
#‘date_time’이라는 컬럼을 date형식의 data로 불러 읽어오라는 명령 
# 위와 같이 컬럼을 date형식으로 처리 하고 난 후에야, date_time.dt 라는 메소드가 잘 작동한다. 
```

```python
ufos[ufos["shape"]== "circle"].date_time.dt.date.value_counts().sort_index().plot(label="circle")
ufos[ufos["shape"] == "light"].date_time.dt.date.value_counts().sort_index().plot(label="light")
ufos[ufos["shape"] == "triangle"].date_time.dt.date.value_counts().sort_index().plot(label="triangle")
ufos[ufos["shape"] == "fireball"].date_time.dt.date.value_counts().sort_index().plot(label="fireball")
ufos[ufos["shape"] == "cigar"].date_time.dt.date.value_counts().sort_index().plot(label="cigar")
plt.legend()
#plt.xlim(2010,2018)
```

```python
ufos[ufos["shape"]== "circle"].date_time.dt.year.value_counts().sort_index().plot(label="circle")
ufos[ufos["shape"] == "triangle"].date_time.dt.year.value_counts().sort_index().plot(label="light")
ufos[ufos["shape"] == "fireball"].date_time.dt.year.value_counts().sort_index().plot(label="triangle")
ufos[ufos["shape"] == "cigar"].date_time.dt.year.value_counts().sort_index().plot(label="fireball")
ufos[ufos["shape"] == "formation"].date_time.dt.year.value_counts().sort_index().plot(label="cigar")
plt.legend()
plt.xlim(2010,2018)
```

```python
ufos[ufos["shape"]== ufos['shape']].date_time.dt.year.value_counts().sort_index().plot()
plt.xlim(1970,2020)
```

```python
ufos.date_time.dt.month.value_counts().sort_index().plot(kind='bar')
```

```python
ufos.date_time.dt.year.value_counts().sort_index().plot()
```

```python
months = {1:"Jan",2:"Fed",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"}
fig, axs = plt.subplots(2, 3, figsize=(14, 8))
ufos[ufos.date_time.dt.year == 2014].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[0][0], kind="bar",title=('2014'))
ufos[ufos.date_time.dt.year == 2015].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[0][1], kind="bar",title=('2015'))
ufos[ufos.date_time.dt.year == 2016].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[0][2], kind="bar",title=('2016'))
ufos[ufos.date_time.dt.year == 2017].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[1][0], kind="bar",title=('2017'))
ufos[ufos.date_time.dt.year == 2018].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[1][1], kind="bar",title=('2018'))
ufos[ufos.date_time.dt.year == 2019].date_time.dt.month.value_counts().sort_index().rename(months).plot(ax=axs[1][2], kind="bar",title=('2019'))
```

```python
ufos[ufos["shape"]== "light"].date_time.dt.year.value_counts().sort_index().plot(label="light")
ufos[ufos["shape"] == "fireball"].date_time.dt.year.value_counts().sort_index().plot(label="fireball")
ufos[ufos["shape"] == "circle"].date_time.dt.year.value_counts().sort_index().plot(label="circle")
ufos[ufos["shape"] == "triangle"].date_time.dt.year.value_counts().sort_index().plot(label="triangle")
ufos[ufos["shape"] == "formation"].date_time.dt.year.value_counts().sort_index().plot(label="formation")
plt.legend()
plt.xlim(2000.0,2018.5)
plt.title("UFO Sightings By Shape")
```
